## Lists
Python lists are dynamical array lists. This means that they consist of a memory-contiguous array of pointers to objects (instead of a non-contiguous linked list of object containers). In order to be able to quickly append to a list, it typically allocates an array of pointers larger than necessary.

```python
l = ['zero', 'one', 'two', 'three', 'four']
```

```python
type(l)
```

Watch out, as for most general purpose programming languages, in Python, counting and indexing generally start at 0. (This is contrary to e.g. matlab and fortran, but in line with C, Java, etc.)

```python
l[2]  # the 3rd element of the list
```

We can index the end of the list by using negative numbers (matlab correspondence `l[end]`)

```python
l[-1], l[-2]
```

## For loops
For loops are generally over items in lists or other iterables. This is contrary to most other languages where an integer counter is the standard way of following the progress of the loop.

```python
l
```

```python
for a in l:
    print("Hello, {}".format(a))
```

If you really need a counter, you can `enumerate` the iterator:

```python
for i, a in enumerate(l):
    print "{counter}. Hello, {item}".format(counter=i, item=a)
```

If you really only want to count, then `range` is the appropriate iterable

```python
for i in range(10): print(i)
```

## Functions
For programming going beyond simple scripts it is essential to encapsulate functionality in functions. They are easy to define:

```python
def double_x(x):
    return 2 * x
```

```python
double_x(4)
```

```python
double_x('abc')
```

```python
double_x([True, False, None])
```

Python is **ducktyped** - *"If it looks like a duck, swims like a duck, and quacks like a duck, then it probably is a duck."* - if the operation `2 * x` is possible, then `double_x` will apply it, no matter what the object is.
